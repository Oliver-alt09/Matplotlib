```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes()
x=np.linspace(1,50,30000)
ax.plot(x,np.sin(x))
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.log(x))
plt.show()
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 100), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 200), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 300), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 400), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 500), color='chartreuse'); # all HTML color names supported
plt.show()
```

```python
plt.plot(x,np.sin(x),linestyle='dashed')
plt.show()
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 10, linestyle='dashed')
plt.plot(x, x + 20, linestyle='dashdot')
plt.plot(x, x + 30, linestyle='dotted');
 # For short, you can use the following codes:
plt.plot(x, x + 40, linestyle='-') # solid
plt.plot(x, x + 50, linestyle='--') # dashed
plt.plot(x, x + 60, linestyle='-.') # dashdot
plt.plot(x, x + 70, linestyle=':'); # dotted
plt.show()
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 10, '--c') # dashed cyan
plt.plot(x,x+20, '-.k') # dashdot black
plt.plot(x, x + 30, ':r'); # dotted red
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-50,50)
plt.ylim(-10,10)
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.axis([-50,50,-2,2])
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
plt.show()
```

```python
plt.plot(x,np.sin(x),label='sin(x)')
plt.axis('equal')
plt.xlabel('X')
plt.ylabel('Sin(x))')
plt.title('A Sine Curve')
plt.legend()
plt.show()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
import random
sizes=[60,30,10,80,40,100,18]
y=np.linspace(1,200,90)
plt.scatter(y, np.sin(y), c=np.random.randint(100,size=len(y)), s=np.random.choice(sizes,size=len(y)), alpha=0.8,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
plt.errorbar(y,np.sin(y), yerr=0.1,fmt='.-.k')
plt.show()
```

```python
plt.errorbar(y,np.sin(y),yerr=0.3, fmt='o' , color='green', ecolor='orange', elinewidth=4, capsize=0)
plt.show()
```

```python
x1,x2=np.meshgrid(y,np.tan(y))
plt.contour(x1,x2,x1*np.cos(y)+x2,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import numpy as np
data=np.random.randn(30000)
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=20, density=True, alpha=0.7, histtype='stepfilled',color='darkslategray',edgecolor='steelblue')
plt.show()
```

```python
a=np.random.normal(0,0.5,20000)
b=np.random.normal(-2,3,20000)
c=np.random.normal(2,1,20000)
args = dict(histtype='stepfilled', alpha=0.5, density=True, bins=50)
plt.hist(a, **args,label='a')
plt.hist(b, **args,label='b')
plt.hist(c, **args,label='c')
plt.legend(loc='center left',frameon=True , ncol=3, borderpad=1.5, shadow=True, framealpha=0.9, fancybox=True)
plt.show()
```

```python
plt.plot(a,b,c)
plt.legend(['a','b','c'])
plt.show()
```

```python
mean=[1,0]
cov=[[1,5],[2,5]]
x,y=np.random.multivariate_normal(mean, cov, 30000).T
plt.hist2d(x,y, bins=50, cmap='viridis')
cb=plt.colorbar()
cb.set_label('Bins')
plt.show()
```

```python
 np.histogram2d(x, y, bins=30)
```

```python
plt.hexbin(x,y, gridsize=30, cmap='cubehelix')
plt.colorbar(label='count in bin')
plt.show()
```
